## **1. Logarithmic Plotting**

### **a)**
Consider two mathematical functions, $y_1 = x$, and $y_2 = x^2$. 

Create appropriate vectors to represent these functions, for an x-range of 0 to 1

````python
x = linspace(0, 1, 1000)
y1 = x
y2 = x**2
````

### **b)**
Plot both functions on a single ﬁgure. Use a command like 
```python 
fig1=figure()
```
before plotting to set up a speciﬁc ﬁgure window. 

Re-adjust the axis ranges manually (within your plotting script).

Include a title, axis labels, and legends.

The line and marker styles, colors, and font size are up to you.

````python
# figure size
fig1 = figure(figsize=(8, 6))

# Graph information
title('Part B: Linear and Quadratic')
xlabel('$x$')
ylabel('$y$')

grid()
plot(x, y1, label='$y = x$')
plot(x, y2, label='$y = x^2$')

legend()

show()
````

### **c)**
Set up a second ﬁgure window and plot $\log(y1)$ and $\log(y2)$ vs. $\log(x)$ (base-10
logarithms).

Use ﬁgure decoration as above. What happened when you plotted the logarithm of zero?

````python
f1 = log10(y1)
f2 = log10(y2)
f3 = log10(x)

fig2 = figure(figsize=(8, 6))

# Graph information
title('Part C: Logarithmic')
xlabel('$x$')
ylabel('$y$')

grid()
plot(x, f1, label='$y = \log_{10} y1$')
plot(x, f2, label='$y = \log_{10} y2$')
plot(x, f3, label='$y = \log_{10} x$')

legend()

show()
````

We get a `RuntimeWarning: divide by zero encountered in log10` error when trying to plot the logarithm of `0`

### **d)**
Set up a third ﬁgure and make a log–log plot as before, but this time using the 
```python
loglog()
```
 command.

````python
fig3 = figure(figsize=(8, 6))

# Graph information
title('Part D: Log-Log')
xlabel('$x$')
ylabel('$y$')

grid()
loglog(x, y1, label='$y = \log x$')
loglog(x, y2, label='$y = x^2$')
loglog(x, x, label='$y = x$')

legend()

show()
````

### What is `loglog()`?

While the normal `log()` function just returns the logarithmic value of the numbers, the `loglog()` function creates a plot with both axes scaled logarithmically, which helps to clearly see relationships in data that grows or shrinks exponentially.



## **2) List Slicing**

Slice the string
```python
s=’seehemewe
``` 
(using indices) to produce the following sub-strings:

- `a. ’see’`
- `b. ’he’`
- `c. ’me’`
- `d. ’we’`
- `e. ’hem’`
- `f. ’meh’`
- `g. ’wee’`


````python
s = 'seehemewe'
print(s[:3]) # a
print(s[3:5]) # b
print(s[5:7]) # c
print(s[7:9]) # d
print(s[3:6]) # e
print(s[5:2:-1]) # f
print(s[7::-3]) # g
````

### **Python Slicing Syntax:**
The general syntax for slicing is:

```python
string[start:stop]
```
- **`start`**: The index where the slice begins (inclusive, or includes the **`start`**).
- **`stop`**: The index where the slice ends (exclusive, or does not include the **`stop`**).

You can also specify a **step** to control the interval between the characters, like `s[start:stop:step]`
