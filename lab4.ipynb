## 4.1 Compound Interest



The following Python code attempts to calculate the balance of a savings account with an annual
interest rate of 5% after 4 years, assuming that the account starts with a balance of $100:
```python
>>> balance = 100
>>> def add_interest(balance, rate):
... balance += balance * rate / 100
...
>>> for year in range(4):
...     add_interest(balance, 5)
...     print('Balance after year {}: ${:.2f}'.format(year+1, balance))
...
Balance after year 1: $100.00
Balance after year 2: $100.00
Balance after year 3: $100.00
Balance after year 4: $100.00
```

a) First, explain why this code does not work

The code does not work because it is missing the `return` statement for `add_interest` so it isn't being saved into a variable. In addition, in the for-loop, even if there was a return statement in the function, it isn't reassigning `balance` the value that results from `add_interest` (e.g. `balance = add_interest(balance, 5)`)

b) Fix the above code or provide an alternative Python code that does work as intended. Be sure to include your new code and its output with your lab submission.


````python
def add_interest(balance, rate):
    balance += balance * rate / 100
    return balance

balance = 100
for year in range(4):
    balance = add_interest(balance, 5)
    print('Balance after year {}: {:.2f}'.format(year + 1, balance))
````

## Range and Height of a Projectile

The range of a projectile launched at an angle α and speed v on a flat surface is given by
$$
R = \frac{v^2\sin(2\alpha)}{g}
$$
where $g=9.8\frac m {s^2}$ is the acceleration due to Earth's gravity

The maximum height that the projectile achieves is given by 
$$
H = \frac{v^2\sin(2\alpha)}{2g}
$$

(neglecting air resistane and the curvature and rotation of the Earth)

Write a function in Python to calculate and return the range $R$ and maximum height $H$ of a
projectile, taking $\alpha$ and $v$ as arguments. 

Test your code using the values $\alpha = 30\degree$ and $v = 10\frac m {s}$.

````python
# function for range 
def projRange(alpha, velocity):
    g = 9.81 # acceleration due to gravity
    return (velocity**2 * np.sin(2*alpha))/g

# function for height
def projHeight(alpha, velocity):
    g = 9.81 # acceleration due to gravity
    return (velocity**2 * np.sin(2*alpha))/2*g
````

````python
alpha = np.pi/6 # 30 degrees
velocity = 10 # m/s

print(f'Range: {projRange(alpha, velocity):.2f} m')
print(f'Height: {projHeight(alpha, velocity):.2f} m')
````

## 4.3 Position, Velocity, and Acceleration of a Particle 

The position of $x$ as a function of time of a particle that moves in one dimension is given by
$$
x(t) = -0.2t^3 + 0.5t^2 + 4t - 22
$$
where $x$ has units of meters and $t$ units of seconds. The velocity $v(t)$ of the particle is determined
by the derivative of $x(t)$ with respect to $t$, and the acceleration $a(t)$ is determined by the derivative
of $v(t)$ with respect to t (i.e., $v(t) = \frac{dx}{dt}$ and $a(t) = \frac{dv}{dt})$

a) First, derive the analytic expressions for the velocity and acceleration of the particle, and deﬁne Python functions for these, as well as for the position x (three functions in total)

````python
def position(time):
    return (-0.2*(time**3)) + (0.5*(time**2)) + (4*time) - 22
def velocity(time):
    return (-0.6*(time**2)) + (1*time) + 4
def acceleration(time):
    return (-1.2*time) + 1
````

b) Next, add code that prompts the user to input a value of $t$, and then calculates and displays the corresponding values of $x$, $v$, and $a$ for this input value of $t$. Use the functions you deﬁned in part (a), and properly formatted output

````python
t = float(input('Enter time: '))
print(f'Position: {position(t):.2f} m')
print(f'Velocity: {velocity(t):.2f} m/s')
print(f'Acceleration: {acceleration(t):.2f} m\s^2')
````

c) Finally, use your Python functions to make plots of the position, velocity, and acceleration as a function of time for the time interval 0 ≤ t ≤ 8. Use plot titles and label the axes appropriately with the correct units. Hint: enclose mathematical expressions in \$ symbols to get nicer
formatting. For example: `xlabel(‘$x$ [m]’)`.

````python
x = linspace(0, 8, 1000)
title("Position, Velocity, Acceleration vs Time")
xlabel("$x$ (s)")
ylabel("$x(t)$ (m), $v(t)$ (m/s), $a(t)$ (m/s^2)")
plot(x, position(x), label="Position")
plot(x, velocity(x), label="Velocity")
plot(x, acceleration(x), label="Acceleration")
legend()
autoscale()
````
