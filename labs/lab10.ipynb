# **Lab 10: Home Run Derby**

```python
from pylab import *
from random import uniform
```

## Case 1: Hitting the Ground

```python
# constants/variables 
g = 9.81 # m/s^2
v_0 = float(input("Enter the initial velocity: ")) # m/s
theta_0 = float(input("Enter the initial angle: ")) # degrees
is_on_wall: bool = False
is_past_wall: bool = False
hit_ground: bool = False
wall_length = 100 # m
wall_height = 8 # m

print(f'Initial velocity: {v_0} m/s | Initial angle: {theta_0} degrees')

# function for duration of projectile motion
def t_flight(v_0, theta_0) -> float:
    return 2*v_0*sin(radians(theta_0))/g

print(f'Time of flight {t_flight(v_0, theta_0):.2f} seconds')
```

```python
# vector of time values
t = linspace(0, t_flight(v_0, theta_0), 100)

# x and y compononents of the projectile
x = v_0*t*cos(radians(theta_0))
y = v_0*t*sin(radians(theta_0)) - 0.5*g*t**2

for i in range(len(t)):
    if y[i] < 0:
        y[i] = 0
        hit_ground = True
        break
    if x[i] >= wall_length: 
        is_past_wall = True
        x[i:] = wall_length
        y[i:] = y[i]
        break
    if x[i] == wall_length and y[i] <= wall_height:
        is_on_wall = True
        break
    print(f'Time: {t[i]:.2f}s | x: {x[i]:.2f}m | y: {y[i]:.2f}m')

# check in which state the ball is in
if is_past_wall:
    print("Home run!!")
elif is_on_wall:
    print("The baseball hit the wall!")
else:
    print("The baseball hit the ground.")
```

```python

fig = figure(figsize=(8, 6))
ax = fig.add_subplot(111, aspect='equal')

# title 
title('Home Run Derby: $y(t)$ vs $x(t)$')
# label the axes
xlabel('x(t) (m)')
ylabel('y(t) (m)')

# set the limits of the axes
xlim(0, wall_length + 10) # padding of 10
ylim(0, y.max() + 10)

# plot vertical wall at x = 100 with height 8 and scale the axes
plot([wall_length, wall_length], [0, wall_height], 'r-')
# plot the trajectory
plot(x, y, 'y-')
```

## Case 3: On the Wall

```python
# constants/variables 
g = 9.81 # m/s^2
is_on_wall: bool = False
is_past_wall: bool = False
hit_ground: bool = False
wall_length = 100 # m
wall_height = 8 # m

```

```python
while not is_on_wall:
    # random initial velocity and angle
    v_0, theta_0 = uniform(1., 100.), uniform(1., 89.) # m/s, degrees
    print(f'Initial velocity: {v_0:.4f} m/s | Initial angle: {theta_0:.4f} degrees')

    # function for duration of projectile motion
    def t_flight(v_0, theta_0) -> float:
        return 2*v_0*sin(radians(theta_0))/g

    print(f'Time of flight {t_flight(v_0, theta_0):.4f} seconds')

    # vector of time values
    t = linspace(0, t_flight(v_0, theta_0), 100)

    # x and y compononents of the projectile 
    x = v_0*t*cos(radians(theta_0))
    x = x[x <= wall_length] # only consider x values that are less than or equal to the wall length
    t_wall = t[:len(x)] # time when the ball reaches the wall
    y = v_0*t_wall*sin(radians(theta_0)) - 0.5*g*t_wall**2

    print(f'x {x[-1]:.2f} | y {y[-1]:.2f}')
    
    # check if the ball is on the wall (tolerance of 0.1)
    if 0 < y[-1] <= wall_height and wall_length - 0.1 <= x[-1] <= wall_length:
        is_on_wall = True

print("The baseball hit the wall!")
```

```python

fig = figure(figsize=(8, 6))
ax = fig.add_subplot(111, aspect='equal')

# title 
title('Home Run Derby: $y(t)$ vs $x(t)$')
# label the axes
xlabel('x(t) (m)')
ylabel('y(t) (m)')

# set the limits of the axes
xlim(0, wall_length + 10) # padding of 10
ylim(0, y.max() + 10)

# plot vertical wall at x = 100 with height 8 and scale the axes
plot([wall_length, wall_length], [0, wall_height], 'r-')
# plot the trajectory
plot(x, y, 'y-')
```
