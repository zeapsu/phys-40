## Dropping lead balls

```python
from pylab import *
```

```python
# constants
z0 = 9  #m
g = 9.81  #m/s^2
dt = 0.01  #s

# initialize time and height arrays
t = [0] 
z = [z0] 

# simulation loop
while z[-1] > 0:
    t_new = t[-1] + dt
    z_new = z0 - 0.5 * g * t_new ** 2
    if z_new < 0:
        z_new = 0  # ensure the ball doesn't go underground lol
    t.append(t_new)
    z.append(z_new)
    print(f'Time: {t_new:.2f} s, Height: {z_new:.2f} m')

# calculate velocity
vz = [-g * time for time in t]  # expected velocity: vz = -g * t
vz_approx = [(z[i] - z[i-1]) / dt for i in range(1, len(z))]  # approximate velocity from position: vz_approx = ∆z/∆t
vz_approx.insert(0, 0)  # initial velocity is 0

# plotting the height versus time
figure(figsize=(10, 5))

# plot height vs time
subplot(1, 2, 1)
plot(t, z, label="Height z(t)")
xlabel('Time (s)')
ylabel('Height (m)')
title('Height vs Time')
grid(True)

# plot velocity vs time
subplot(1, 2, 2)
plot(t, vz, label="Expected Velocity v(t) = -gt", linestyle='--')
plot(t, vz_approx, label="Calculated Velocity from ∆z/∆t")
xlabel('Time (s)')
ylabel('Velocity (m/s)')
title('Velocity vs Time')
grid(True)
legend()

tight_layout()
show()

```
