# **Lab 12: Exoplanets**



```python
from pylab import *
import pandas as pd
```

```python
# Read in data from the csv
data = pd.read_csv('exoplanets.csv')
data

```

```python
# constants for conversion
earthRadius = 6371 # in km
au = 1.5*10**8 # in km

# make vectors for each field
planets = array(data['Planet']) 
orbitalPeriods = array(data['Orbital Period']) # days
orbitalPeriods /= 365.25 # years
planetRadius = array(data['Planet Radius']) # R_\oplus
planetMass = array(data['Planet Mass']) # M_\oplus
```

Note that Planet radius and Planet mass are in terms of $R_\oplus = 6371\;km$, $M_\oplus = 5.972\times 10^{24}\;kg$ respectively

```python
# turn everything into mks units
planetRadius *= 6371 # in km
planetRadius
```

```python
planetMass *= 5.972*10**24 # in kg
planetMass
```

## Part a:

```python
# Kepler's third law
# returns the semi-major axis length 'a' of ea planet's orbit in AU
def keplerThirdLaw(orbitalPeriod):
    return (orbitalPeriod**(2/3))

semiMajorAxisLength = keplerThirdLaw(orbitalPeriods) # in AU
semiMajorAxisLength
```

## Part b:

```python
for i, item in enumerate(semiMajorAxisLength):
    print(f'Planet {planets[i]}: semi-major axis = {item:.3} AU = {item * (1.5*10**8):.3} km')
```

## Part c:

```python

def drawOrbit(x0: int, y0: int, r: float, numPoints = 100) -> (float, float):
    theta = linspace(0, 2*pi , numPoints)
    x = r * cos(theta) + x0
    y = r * sin(theta) + y0
    return x, y

fig = figure(figsize=(8, 8))
ax = fig.add_subplot(111)
# rcParams['lines.markersize'] = 1

# plot information
title("Orbits of Trappist-1 Planets")
xlabel('distance ($km$)')
ylabel('distance ($km$)')

# plot the orbit for each planet
for radius in planetRadius:
    # print(f'Radius is {radius:.3} km')
    x, y = drawOrbit(0, 0, radius)
    plot(x, y, linestyle='dotted', linewidth=1, color='k', label=f'$x^2 + y^2 = {radius:.3}^2$')

starRadius = (12.45 * earthRadius) / 35
x, y = drawOrbit(0, 0, starRadius)
plot(x, y, 'r-', label=f'$x^2 + y^2 = {starRadius:.3}^2$')
autoscale(axis='y')

# draw a circle to represent planet in each orbit

for radius in planetRadius:
    x, y = drawOrbit(radius, 0, (radius / earthRadius) * 100)
    plot(x, y, '-', label=f'$x^2 + y^2 = {radius:.3}^2$')
```
