### Add a legend to the lower-right corner

```python

# label the x and y axes
xlabel('$x$ (time)',fontsize=14)
ylabel('$y$ (chickens)',fontsize=14)

# add a title
title('Chickens vs. time', fontsize=16)

# plot the data
plot(x, y, 'k-')
plot(x, y, 'bx')

# plot a point
text(0.1, -1.4, 'A very important point')
plot(4 , -1.41, 'go')

# add legend to lower right, with no frame
legend(['sin(x)+cos(x)'], loc='lower right', frameon=True)
```

### Turn on grid lines

```python

# label the x and y axes
xlabel('$x$ (time)',fontsize=14)
ylabel('$y$ (chickens)',fontsize=14)

# add a title
title('Chickens vs. time', fontsize=16)


# plot the data
plot(x, y, 'k-')
plot(x, y, 'bx')

# plot a point
text(0.1, -1.4, 'A very important point')
plot(4 , -1.41, 'go')

# add legend to lower right, with no frame
legend(['sin(x)+cos(x)'], loc='lower right', frameon=True)

# show grid 
grid(True)
```

### Modify the following command to set the plot font size to look good:
`rcParams.update({'font.size': XX})`

```python

# label the x and y axes
xlabel('$x$ (time)',fontsize=14)
ylabel('$y$ (chickens)',fontsize=14)

# add a title
title('Chickens vs. time', fontsize=16)

# change the font size
rcParams.update({'font.size': 10})

# plot the data
plot(x, y, 'k-')
plot(x, y, 'bx')

# plot a point
text(0.1, -1.4, 'A very important point')
plot(4 , -1.41, 'go')

# add legend to lower right, with no frame
legend(['sin(x)+cos(x)'], loc='lower right', frameon=True)

# show grid 
grid(True)
```

### Save your plot as an image ﬁle (use savefig()), and upload it to Canvas
