# Example of a 3D Plot

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig1 = figure()
ax = fig1.add_subplot(projection='3d')
x = linspace(-10.0,10.0,20)
y = copy(x)
X,Y = meshgrid(x,y)
Z = X**2 + Y**2
ax.plot_surface(X,Y,Z,cmap=cm.magma)
```

# Demo of Stylistic Guidelines for Plotting 

* Label axes, including units
* Use legends
* Use legible fonts and size
* Clarity, simplicity

```python
from pylab import *   # numerical and plotting libraries

x_mod = linspace(0,10,100) # x-values for model: use fine spacing for smooth curve
y_mod = x_mod**2 

x_data = array([1.0,2.0,5.0,7.0])
y_data = array([1.6,3.0,20.0,52.0])
dy = array([1.5,1.7,2.5,3.5]) # measurement uncertainties for y

rcParams.update({'font.size': 18})
fig1=figure(figsize=(9,7))
ax=fig1.add_subplot(111)
plot(x_mod,y_mod,'k-',label='model')
errorbar(x_data,y_data,dy,fmt='go',markersize=10,label='data')
xlim(0,8)  # set axis ranges to focus in on area of interest (where data are)
ylim(0,60)
handles, labels = ax.get_legend_handles_labels() # extra command to set up reverse legend order
ax.legend(handles[::-1],labels[::-1],loc='best',numpoints=1,fontsize=17) 
xlabel('$x$ [units]',fontsize=22)
ylabel('$y$ [units]',fontsize=22)
title('informative title of plot') # not necessary if caption says the same thing
tick_params('both', length=10, width=1.2, which='major') # bigger axis ticks
#savefig('demo_plot.png') #Uncomment beginning of this line to save figure as a .png file
```
